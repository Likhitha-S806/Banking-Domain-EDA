**Categorical Analysis**

```python
#  we want to categorize a continuous variable (Estimated Income) into income bands (Low, Med, High) using pandas.cut()
bins = [0, 100000, 300000, float('inf')] # Defines intervals for categorization:
#0 to <100000 → Low, 100000 to <300000 → Med, 300000 and above → High
labels = ['Low', 'Med', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
# pd.cut(...)Splits the Estimated Income into the specified bins and assigns labels.
```

```python
df['Income Band'].value_counts()
df['Income Band'].value_counts().plot(kind='bar',color = 'green')
```

```python
# Examine the distribution of unique cataegories in categorical columns
categorical_cols = df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns

for col in categorical_cols:
  print(f"Value Counts for '{col}':")
  display(df[col].value_counts())
```

**Numerical Analysis**

```python
df.columns
```

```python
#  Grid layout (multiple plots in one figure)
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 
                  'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 
                  'Foreign Currency Account', 'Business Lending']

plt.figure(figsize=(15, 12))
for i, col in enumerate(numerical_cols):
    plt.subplot(4, 3, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.tight_layout() #avoids overlapping titles and axes.
plt.show()
```

```python
#  Separate plots (one at a time)
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 
                  'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 
                  'Foreign Currency Account', 'Business Lending']

for col in numerical_cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.show()
```

**HeatMaps for Correlation**

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
# annot=True: Shows the actual correlation values in the boxes.
# cmap='crest': Uses a pleasant blue-green color scale.
plt.title("Correlation Matrix") 
plt.show()
```

## Insights of EDA:
The strongest **positive correlations** are observed between **"Bank Deposits" and "Checking Accounts", "Saving Accounts", and "Foreign Currency Account". This pattern suggests a strong financial consistency** i.e customers who maintain high balances in one type of account tend to exhibit similar behavior across other account types, reflecting a well-rounded and stable banking relationship.
